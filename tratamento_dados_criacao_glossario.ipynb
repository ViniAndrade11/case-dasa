```python
#Importando bibliotecas

import pandas as pd
import numpy as np
```

```python
#Ler arquivo e amostra inicial dos dados

df = pd.read_csv('base_headcount.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
#Visualizar estrutura da base

df.shape
```

```python
#Verificar valores se existe valores duplicados em employee_id
df['employee_id'].duplicated().any()
```

```python
#Verificar tipo de dados

df.info()
```

```python
#Verificar colunas com valores nulos

df.isnull().sum()
```

Todas as colunas estão com informação preenchida exceto as colunas que poderiam não ter informações

```python
#Verificar os dados em cada coluna

for coluna in df.columns:
    print(df[coluna].value_counts())
    print('-'*100)
```

Coluna 'hire_date', 'termination_date', 'last_promotion_date', 'last_training_date' são colunas de datas e estão em formato object, proximo passo é transformar em tipo data

```python
#Transformar as colunas de data em formato data

colunas_data = ['hire_date', 'termination_date', 'last_promotion_date', 'last_training_date']
for coluna in colunas_data:
    df[coluna] = pd.to_datetime(df[coluna], errors='coerce')
```

```python
#Verificar se as colunas estão na formatação correta

df[['hire_date', 'termination_date', 'last_promotion_date', 'last_training_date']].info()
```

```python
#Descrição estatisca inicial para identificar desvios

df.describe()
```

Nenhuma inconsistência encontrada.

Unica coluna que poderia ter valores negativos seria bank_hours

Não há nenhuma coluna com valor máximo que não faria sentido, exemplo, number_of_dependents ter um valor igual a 20

Trativa que será feita para melhorar o consumo dos dados é dividir por 100 os dados da coluna bonus_percentage para estar entre 0 e 1

```python
#Dividindo a coluna bonus_percentage para variar em 0 e 1

df['bonus_percentage'] = df['bonus_percentage'] / 100
df['bonus_percentage'].describe()
```

Verificar se existe alguma inconsistência nas colunas em formato de data

```python
#Verificar se existe datas de contratação posterior as datas de saída do colaborador

df['diferenca_dias'] = (df['termination_date'] - df['hire_date']).dt.days
df['diferenca_dias'].describe()
```

```python
df.drop('diferenca_dias', axis=1, inplace=True)
df['diferenca_dias'] = (df['last_promotion_date'] - df['hire_date']).dt.days
df['diferenca_dias'].describe()
```

```python
df.drop('diferenca_dias', axis=1, inplace=True)
df['diferenca_dias'] = (df['last_training_date'] - df['hire_date']).dt.days
df['diferenca_dias'].describe()
```

Existe dados onde o last_training_date é anterior ao hire_date

```python
#Visualizar os dados onde last_training_date é anterior ao hire_date

df_filtrado = df.query('diferenca_dias < 0')
df_filtrado.head()
```

```python
#Verificar quantos registros isso acontece
df_filtrado.shape
```

```python
df_filtrado.describe()
```

Como pode ter sido um erro nos dados e os demais dados parecem coerentes, será feito a substituição para valores nulos

```python
#Substituir por nulo quando a diferença for negativa
df.loc[df['diferenca_dias'] < 0, 'last_training_date'] = pd.NaT
```

```python
#Verificar se deu certo
df[df['diferenca_dias'] < 0][['last_training_date', 'diferenca_dias']].value_counts()
```

```python
df.drop('diferenca_dias', axis=1, inplace=True)
```

```python
#Verificar se todos os colaboradores como False em is_active estão com termination_date preenchido

df_filtrado = df.query('is_active == "False"')
```

```python
df_filtrado['termination_date'].isnull().sum()
```

Todos os colaboradores como False em is_active estão com termination_date preenchido
